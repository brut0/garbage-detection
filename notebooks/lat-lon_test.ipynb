```python
import numpy as np
import pandas as pd
import re
from geopy.geocoders import Nominatim
```

```python
df = pd.read_csv('./Геолокация камер - cctrt_cams.csv')
```

```python
df
```

```python
df['Адрес установки камеры'].apply(lambda x: x.replace('БД; ', '').\
                                   replace('Безопасный город; ', '').\
                                   replace('ТКО', '').\
                                   replace('ТБО', '').\
                                   replace('парковка', '').\
                                   replace('подъезд', '').\
                                   replace(';', ',')
                                  )
```

```python
df['addr'] = df['Адрес установки камеры'].apply(lambda x: x.replace('БД; ', '').\
                                   replace('Безопасный город; ', '').\
                                   replace('Безопасный, ', '').\
                                   replace('ТКО', '').\
                                   replace('ТБО', '').\
                                   replace(' парковка', '').\
                                   replace(' подъезд', '').\
                                   replace('г. ', '').\
                                   replace('Казань ', 'Казань, ').\
                                   replace(' Казань', 'Казань,').\
                                   replace(' Двор прилег территория', '').\
                                   replace(' дворовая проезжая часть', '').\
                                   replace(' Дворовая проезжая часть', '').\
                                   replace('овая проезжая часть', '').\
                                   replace(' дет. площадка', '').\
                                   replace(' двор', '').\
                                   replace(' Парковка', '').\
                                   replace(' Двор прилегающая территория', '').\
                                   replace(' вни перпендикулярно дому', '').\
                                   replace(' обзор вдоль дома', '').\
                                   replace(' Торец балконого выступа', '').\
                                   replace(' перекресток', '').\
                                   replace(' эт.', '').\
                                   replace(' 2эт..', '').\
                                   replace(' Двор', '').\
                                   replace(' въезд во', '').\
                                   replace(' въезд', '').\
                                   replace(' (оз. Лебяжье) – подземный переход (2)', '').\
                                   replace(' (1)', '').\
                                   replace(' (2)', '').\
                                   replace(';', ',')
                                  )
```

```python
latitude = []
longitude = []
geolocator = Nominatim(user_agent="ny_explorer")
for addr in df['addr']:
    while True:
        try:
            #print(addr)
            location = geolocator.geocode(addr)
            #print(location)
            latitude.append(location.latitude)
            longitude.append(location.longitude)
            #print(location.latitude)
        except Exception as e:
            print(addr)
            print(e)
            latitude.append(0.0)
            longitude.append(0.0)
            break
        else:
            break
```

```python
geolocator.geocode("Казань, Проспект Победы д. 139")
```

```python
df['Latitude'] = latitude
df['Longitude'] = longitude
```

```python
df = df[['Адрес установки камеры', 'ID папки', 'Latitude', 'Longitude']]
df
```

```python
df.to_csv('lat-lon_test.csv')
```
